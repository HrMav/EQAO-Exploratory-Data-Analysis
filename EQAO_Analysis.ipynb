# EQAO Schools Analysis

The Education Quality and Accountability Office (EQAO) is an arm's-length agency that provides information about student achievement in Ontario schools and school boards. This analysis delves deep into the EQAO dataset, which encompasses achievements across various grades in key areas like reading, writing, and mathematics.

The primary objective of our exploration is to understand the performance patterns of students across different grades. 

This report encapsulates the findings from our data analysis, providing educators, policymakers, and stakeholders a comprehensive understanding of student achievements in the Ontario school system.

```python
import pandas as pd

eqao_data = pd.read_csv("EQAO.csv")

eqao_data.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set_style("whitegrid")


# Filter out rows with zero values for Grade 3 achievements
filtered_data = eqao_data[(eqao_data['% Grade 3 Stds Achieving the Prov Standard in Reading'] > 0) & 
                          (eqao_data['% Grade 3 Stds Achieving Prov Standard Writing'] > 0) &
                          (eqao_data['% Grade 3 Stds Achieving Prov Standard Mathematics'] > 0)]

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

# Plot the distributions
sns.histplot(filtered_data['% Grade 3 Stds Achieving the Prov Standard in Reading'], kde=True, ax=axes[0], bins=30)
axes[0].set_title('Grade 3 Reading Achievement (Excluding Zeros)')
axes[0].set_xlabel('% Achieving Provincial Standard')
axes[0].set_ylabel('Number of Schools')

sns.histplot(filtered_data['% Grade 3 Stds Achieving Prov Standard Writing'], kde=True, ax=axes[1], bins=30)
axes[1].set_title('Grade 3 Writing Achievement (Excluding Zeros)')
axes[1].set_xlabel('% Achieving Provincial Standard')
axes[1].set_ylabel('Number of Schools')

sns.histplot(filtered_data['% Grade 3 Stds Achieving Prov Standard Mathematics'], kde=True, ax=axes[2], bins=30)
axes[2].set_title('Grade 3 Mathematics Achievement (Excluding Zeros)')
axes[2].set_xlabel('% Achieving Provincial Standard')
axes[2].set_ylabel('Number of Schools')

plt.tight_layout()
plt.show()
```
